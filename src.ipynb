```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
coin_tosses = np.random.binomial(n=1,p=0.5,size=100000000)
print(coin_tosses)
frequency = np.count_nonzero(coin_tosses)  
print(frequency)
relative_frequency = frequency/coin_tosses.shape[0]
print(relative_frequency)
```

```python
coin_tosses = np.random.binomial(n=5,p=0.5,size=100000000)
print(coin_tosses)
frequency = np.count_nonzero(coin_tosses ==2) 
print(frequency)
relative_frequency = frequency/coin_tosses.shape[0]
print(relative_frequency)
```
